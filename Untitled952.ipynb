```python
import os
import re
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, r2_score, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# ===============================
# 1. PATH SETTINGS
# ===============================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Regional Office Accessibility"
DATASET_PATH = os.path.join(BASE_DIR, "contact_details_of_regional_branch_offices_of_tiic_31_12_2019.csv")

os.makedirs(BASE_DIR, exist_ok=True)

print("Loading dataset from:", DATASET_PATH)


# ===============================
# 2. LOAD DATASET
# ===============================

df = pd.read_csv(DATASET_PATH)

print("\nDataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())


# ===============================
# 3. CLEAN COLUMN NAMES
# ===============================

df.columns = (
    df.columns
    .str.strip()
    .str.replace("/", "_", regex=False)
    .str.replace(" ", "_")
    .str.replace(".", "", regex=False)
)

print("\nCleaned Columns:")
print(df.columns.tolist())


# ===============================
# 4. BASIC PREPROCESSING
# ===============================

df = df.fillna("Unknown")

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

# Convert pincode to numeric
df["Pincode"] = pd.to_numeric(df["Pincode"], errors="coerce")
df["Pincode"] = df["Pincode"].fillna(df["Pincode"].median())


# ===============================
# 5. FEATURE ENGINEERING
# ===============================

def extract_district(row):
    text = f"{row.get('Address1','')} {row.get('Address2','')} {row.get('Address3','')}"
    text = text.lower()

    districts = [
        "chennai", "coimbatore", "madurai", "salem", "trichy", "tiruchirappalli",
        "tirunelveli", "vellore", "erode", "thanjavur", "thoothukudi",
        "kanchipuram", "cuddalore", "dindigul", "karur", "namakkal",
        "virudhunagar", "sivagangai", "ramanathapuram", "nagapattinam",
        "tiruppur", "krishnagiri", "dharmapuri", "theni", "ariyalur",
        "perambalur", "pudukkottai", "kanyakumari", "nilgiris"
    ]

    for d in districts:
        if d in text:
            return d.title()

    return "Other"


def office_type(name):
    name = name.lower()
    if "regional" in name:
        return "Regional"
    elif "branch" in name:
        return "Branch"
    elif "satellite" in name:
        return "Satellite"
    else:
        return "Office"


def count_phone_numbers(phone):
    phone = str(phone)
    numbers = re.findall(r"\d{5,}", phone)
    return len(numbers)


def has_email(email):
    email = str(email).lower()
    return 1 if "@" in email else 0


df["District"] = df.apply(extract_district, axis=1)
df["Office_Type"] = df["Names_of_Regional___Branch___Satellite_Offices"].apply(office_type)
df["Phone_Count"] = df["Phone_Number"].apply(count_phone_numbers)
df["Has_Email"] = df["Email_Id"].apply(has_email)

# Synthetic demand score because original dataset has only office/contact data
np.random.seed(42)

df["Estimated_Business_Density"] = np.random.randint(50, 500, size=len(df))
df["Estimated_MSME_Count"] = np.random.randint(100, 2000, size=len(df))
df["Estimated_Distance_Index"] = np.random.uniform(1, 50, size=len(df)).round(2)

df["Service_Demand_Score"] = (
    df["Estimated_Business_Density"] * 0.35 +
    df["Estimated_MSME_Count"] * 0.45 -
    df["Estimated_Distance_Index"] * 2 +
    df["Phone_Count"] * 10 +
    df["Has_Email"] * 20
)

df["Demand_Category"] = pd.qcut(
    df["Service_Demand_Score"],
    q=3,
    labels=["Low", "Medium", "High"]
)

print("\nFeature Engineering Completed")
print(df[[
    "District",
    "Office_Type",
    "Phone_Count",
    "Has_Email",
    "Estimated_Business_Density",
    "Estimated_MSME_Count",
    "Estimated_Distance_Index",
    "Service_Demand_Score",
    "Demand_Category"
]].head())


# ===============================
# 6. SELECT FEATURES
# ===============================

feature_cols = [
    "Regional_and_Branch_Code",
    "Names_of_Regional___Branch___Satellite_Offices",
    "Designation",
    "District",
    "Office_Type",
    "Pincode",
    "Phone_Count",
    "Has_Email",
    "Estimated_Business_Density",
    "Estimated_MSME_Count",
    "Estimated_Distance_Index"
]

target_class = "Demand_Category"
target_reg = "Service_Demand_Score"

data = df[feature_cols + [target_class, target_reg]].copy()


# ===============================
# 7. ENCODING
# ===============================

label_encoders = {}

for col in data.select_dtypes(include="object").columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le

target_encoder = LabelEncoder()
data[target_class] = target_encoder.fit_transform(data[target_class].astype(str))

X = data[feature_cols]
y_class = data[target_class]
y_reg = data[target_reg]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ===============================
# 8. TRAIN TEST SPLIT
# ===============================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_class,
    test_size=0.2,
    random_state=42,
    stratify=y_class
)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_scaled,
    y_reg,
    test_size=0.2,
    random_state=42
)


# ===============================
# 9. RANDOM FOREST CLASSIFIER
# ===============================

rf_classifier = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=8
)

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(
    y_test,
    y_pred,
    target_names=target_encoder.classes_,
    output_dict=True
)

print("\nClassification Accuracy:", accuracy)
print("\nConfusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_encoder.classes_))


# ===============================
# 10. RANDOM FOREST REGRESSOR
# ===============================

rf_regressor = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    max_depth=8
)

rf_regressor.fit(X_train_reg, y_train_reg)

reg_pred = rf_regressor.predict(X_test_reg)

r2 = r2_score(y_test_reg, reg_pred)
mae = mean_absolute_error(y_test_reg, reg_pred)

print("\nRegression R2 Score:", r2)
print("Regression MAE:", mae)


# ===============================
# 11. ANN MODEL FOR .h5
# ===============================

num_classes = len(target_encoder.classes_)

ann_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

ann_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=15,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_loss, ann_accuracy = ann_model.evaluate(X_test, y_test, verbose=0)

print("\nANN Test Accuracy:", ann_accuracy)
print("ANN Test Loss:", ann_loss)


# ===============================
# 12. SAVE .h5 MODEL
# ===============================

h5_path = os.path.join(BASE_DIR, "regional_office_accessibility_model.h5")
ann_model.save(h5_path)

print("\nH5 model saved at:", h5_path)


# ===============================
# 13. SAVE .pkl FILE
# ===============================

pkl_data = {
    "random_forest_classifier": rf_classifier,
    "random_forest_regressor": rf_regressor,
    "scaler": scaler,
    "label_encoders": label_encoders,
    "target_encoder": target_encoder,
    "feature_columns": feature_cols,
    "accuracy": accuracy,
    "regression_r2": r2,
    "regression_mae": mae
}

pkl_path = os.path.join(BASE_DIR, "regional_office_accessibility_model.pkl")

with open(pkl_path, "wb") as f:
    pickle.dump(pkl_data, f)

print("PKL file saved at:", pkl_path)


# ===============================
# 14. SAVE YAML CONFIG
# ===============================

yaml_config = {
    "project_name": "Smart Regional Office Accessibility System",
    "dataset_path": DATASET_PATH,
    "output_directory": BASE_DIR,
    "problem_type": {
        "classification": "Demand Category Prediction",
        "regression": "Service Demand Score Prediction"
    },
    "target_columns": {
        "classification_target": target_class,
        "regression_target": target_reg
    },
    "features_used": feature_cols,
    "models": {
        "classification_model": "RandomForestClassifier",
        "regression_model": "RandomForestRegressor",
        "deep_learning_model": "ANN saved as H5"
    },
    "metrics": {
        "classification_accuracy": float(accuracy),
        "ann_accuracy": float(ann_accuracy),
        "ann_loss": float(ann_loss),
        "regression_r2_score": float(r2),
        "regression_mae": float(mae)
    }
}

yaml_path = os.path.join(BASE_DIR, "regional_office_accessibility_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, default_flow_style=False)

print("YAML config saved at:", yaml_path)


# ===============================
# 15. SAVE JSON METADATA
# ===============================

json_metadata = {
    "project_name": "Smart Regional Office Accessibility System",
    "total_records": int(df.shape[0]),
    "total_columns": int(df.shape[1]),
    "original_columns": df.columns.tolist(),
    "engineered_features": [
        "District",
        "Office_Type",
        "Phone_Count",
        "Has_Email",
        "Estimated_Business_Density",
        "Estimated_MSME_Count",
        "Estimated_Distance_Index",
        "Service_Demand_Score",
        "Demand_Category"
    ],
    "classification_labels": target_encoder.classes_.tolist(),
    "classification_accuracy": float(accuracy),
    "ann_accuracy": float(ann_accuracy),
    "ann_loss": float(ann_loss),
    "regression_r2_score": float(r2),
    "regression_mae": float(mae),
    "confusion_matrix": cm.tolist(),
    "classification_report": report,
    "saved_files": {
        "h5_model": h5_path,
        "pkl_model": pkl_path,
        "yaml_config": yaml_path
    }
}

json_path = os.path.join(BASE_DIR, "regional_office_accessibility_metadata.json")

with open(json_path, "w") as f:
    json.dump(json_metadata, f, indent=4)

print("JSON metadata saved at:", json_path)


# ===============================
# 16. SAVE PROCESSED DATASET
# ===============================

processed_csv_path = os.path.join(BASE_DIR, "regional_office_accessibility_processed_dataset.csv")
df.to_csv(processed_csv_path, index=False)

print("Processed dataset saved at:", processed_csv_path)


# ===============================
# 17. SAVE PREDICTIONS CSV
# ===============================

prediction_df = df.copy()

all_class_preds = rf_classifier.predict(X_scaled)
all_class_labels = target_encoder.inverse_transform(all_class_preds)

all_reg_preds = rf_regressor.predict(X_scaled)

prediction_df["Predicted_Demand_Category"] = all_class_labels
prediction_df["Predicted_Service_Demand_Score"] = all_reg_preds.round(2)

prediction_csv_path = os.path.join(BASE_DIR, "regional_office_accessibility_predictions.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("Prediction CSV saved at:", prediction_csv_path)


# ===============================
# 18. FINAL OUTPUT SUMMARY
# ===============================

print("\n======================================")
print("PROJECT FILES GENERATED SUCCESSFULLY")
print("======================================")
print("H5 Model:", h5_path)
print("PKL File:", pkl_path)
print("YAML Config:", yaml_path)
print("JSON Metadata:", json_path)
print("Processed CSV:", processed_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("======================================")
```
