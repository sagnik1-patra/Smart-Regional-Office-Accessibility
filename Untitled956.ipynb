```python
import os
import re
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# ============================================================
# 1. PATH SETTINGS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Regional Office Accessibility"
DATASET_PATH = os.path.join(
    BASE_DIR,
    "contact_details_of_regional_branch_offices_of_tiic_31_12_2019.csv"
)

os.makedirs(BASE_DIR, exist_ok=True)


# ============================================================
# 2. LOAD DATASET
# ============================================================

df = pd.read_csv(DATASET_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())


# ============================================================
# 3. CLEAN COLUMN NAMES
# ============================================================

df.columns = (
    df.columns
    .str.strip()
    .str.replace("/", "_", regex=False)
    .str.replace(".", "", regex=False)
    .str.replace(" ", "_", regex=False)
    .str.replace("__", "_", regex=False)
)

df = df.fillna("Unknown")

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

print("\nCleaned Columns:")
print(df.columns.tolist())


# ============================================================
# 4. AUTO DETECT COLUMNS
# ============================================================

def find_col(possible_words):
    for col in df.columns:
        col_low = col.lower()
        for word in possible_words:
            if word.lower() in col_low:
                return col
    return None


code_col = find_col(["code"])
office_col = find_col(["office", "name"])
designation_col = find_col(["designation"])
address_col = find_col(["address"])
pincode_col = find_col(["pin", "pincode"])
phone_col = find_col(["phone", "mobile", "contact"])
email_col = find_col(["email", "mail"])

print("\nDetected Columns:")
print("Code:", code_col)
print("Office:", office_col)
print("Designation:", designation_col)
print("Address:", address_col)
print("Pincode:", pincode_col)
print("Phone:", phone_col)
print("Email:", email_col)


# ============================================================
# 5. FEATURE ENGINEERING
# ============================================================

if pincode_col:
    df["Pincode_Numeric"] = pd.to_numeric(df[pincode_col], errors="coerce")
    df["Pincode_Numeric"] = df["Pincode_Numeric"].fillna(df["Pincode_Numeric"].median())
else:
    df["Pincode_Numeric"] = np.arange(len(df)) + 600000


def extract_district(text):
    text = str(text).lower()

    districts = [
        "chennai", "coimbatore", "madurai", "salem", "trichy",
        "tiruchirappalli", "tirunelveli", "vellore", "erode",
        "thanjavur", "thoothukudi", "kanchipuram", "cuddalore",
        "dindigul", "karur", "namakkal", "virudhunagar",
        "sivagangai", "ramanathapuram", "nagapattinam", "tiruppur",
        "krishnagiri", "dharmapuri", "theni", "ariyalur",
        "perambalur", "pudukkottai", "kanyakumari", "nilgiris"
    ]

    for district in districts:
        if district in text:
            return district.title()

    return "Other"


if address_col:
    df["District"] = df[address_col].apply(extract_district)
else:
    df["District"] = "Other"


def get_office_type(text):
    text = str(text).lower()

    if "regional" in text:
        return "Regional Office"
    elif "branch" in text:
        return "Branch Office"
    elif "satellite" in text:
        return "Satellite Office"
    else:
        return "General Office"


if office_col:
    df["Office_Type"] = df[office_col].apply(get_office_type)
else:
    df["Office_Type"] = "General Office"


def phone_count(text):
    nums = re.findall(r"\d{5,}", str(text))
    return len(nums)


if phone_col:
    df["Phone_Count"] = df[phone_col].apply(phone_count)
else:
    df["Phone_Count"] = 0


if email_col:
    df["Has_Email"] = df[email_col].apply(lambda x: 1 if "@" in str(x) else 0)
else:
    df["Has_Email"] = 0


np.random.seed(42)

df["Estimated_Business_Density"] = np.random.randint(100, 800, len(df))
df["Estimated_MSME_Count"] = np.random.randint(200, 2500, len(df))
df["Estimated_Distance_Index"] = np.random.uniform(1, 50, len(df)).round(2)
df["Estimated_Population_Coverage"] = np.random.randint(50000, 1000000, len(df))

df["Service_Demand_Score"] = (
    df["Estimated_Business_Density"] * 0.30 +
    df["Estimated_MSME_Count"] * 0.40 +
    df["Estimated_Population_Coverage"] * 0.0002 -
    df["Estimated_Distance_Index"] * 3 +
    df["Phone_Count"] * 15 +
    df["Has_Email"] * 25
)

df["Demand_Category"] = pd.qcut(
    df["Service_Demand_Score"],
    q=3,
    labels=["Low", "Medium", "High"]
)

print("\nFeature Engineering Completed")


# ============================================================
# 6. PREPARE FEATURES
# ============================================================

feature_cols = []

if code_col:
    feature_cols.append(code_col)

if office_col:
    feature_cols.append(office_col)

if designation_col:
    feature_cols.append(designation_col)

feature_cols += [
    "District",
    "Office_Type",
    "Pincode_Numeric",
    "Phone_Count",
    "Has_Email",
    "Estimated_Business_Density",
    "Estimated_MSME_Count",
    "Estimated_Distance_Index",
    "Estimated_Population_Coverage"
]

target_col = "Demand_Category"
reg_target_col = "Service_Demand_Score"

data = df[feature_cols + [target_col, reg_target_col]].copy()

label_encoders = {}

for col in data.columns:
    if data[col].dtype == "object" or str(data[col].dtype) == "category":
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col].astype(str))
        label_encoders[col] = le

target_encoder = label_encoders[target_col]

X = data[feature_cols]
y = data[target_col]
y_reg = data[reg_target_col]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# 7. TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_scaled,
    y_reg,
    test_size=0.25,
    random_state=42
)


# ============================================================
# 8. BASE MODELS
# ============================================================

base_models = {
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=3)
}

results = []
predictions = {}

for model_name, model in base_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results.append({
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_test, y_pred, average="weighted", zero_division=0),
        "F1_Score": f1_score(y_test, y_pred, average="weighted", zero_division=0)
    })

    predictions[model_name] = y_pred


# ============================================================
# 9. PSO OPTIMIZATION - PARTICLE SWARM OPTIMIZATION
# ============================================================

class PSOOptimizer:
    def __init__(
        self,
        population_size=20,
        generations=25,
        inertia=0.7,
        cognitive=1.5,
        social=1.5
    ):
        self.population_size = population_size
        self.generations = generations
        self.inertia = inertia
        self.cognitive = cognitive
        self.social = social

        self.lower_bounds = np.array([50, 3, 2, 1])
        self.upper_bounds = np.array([300, 20, 10, 5])

        self.global_best_position = None
        self.global_best_score = -1
        self.convergence = []

    def initialize_particles(self):
        particles = []
        velocities = []

        for _ in range(self.population_size):
            position = np.array([
                np.random.randint(50, 301),
                np.random.randint(3, 21),
                np.random.randint(2, 11),
                np.random.randint(1, 6)
            ], dtype=float)

            velocity = np.random.uniform(-1, 1, size=4)

            particles.append(position)
            velocities.append(velocity)

        return np.array(particles), np.array(velocities)

    def decode_position(self, position):
        position = np.clip(position, self.lower_bounds, self.upper_bounds)

        return {
            "n_estimators": int(round(position[0])),
            "max_depth": int(round(position[1])),
            "min_samples_split": int(round(position[2])),
            "min_samples_leaf": int(round(position[3]))
        }

    def fitness(self, position):
        params = self.decode_position(position)

        model = RandomForestClassifier(
            n_estimators=params["n_estimators"],
            max_depth=params["max_depth"],
            min_samples_split=params["min_samples_split"],
            min_samples_leaf=params["min_samples_leaf"],
            random_state=42
        )

        model.fit(X_train, y_train)
        pred = model.predict(X_test)

        return accuracy_score(y_test, pred)

    def optimize(self):
        particles, velocities = self.initialize_particles()

        personal_best_positions = particles.copy()
        personal_best_scores = np.array([
            self.fitness(pos) for pos in personal_best_positions
        ])

        best_index = np.argmax(personal_best_scores)
        self.global_best_position = personal_best_positions[best_index].copy()
        self.global_best_score = personal_best_scores[best_index]

        for gen in range(self.generations):
            for i in range(self.population_size):
                r1 = np.random.rand(4)
                r2 = np.random.rand(4)

                velocities[i] = (
                    self.inertia * velocities[i]
                    + self.cognitive * r1 * (personal_best_positions[i] - particles[i])
                    + self.social * r2 * (self.global_best_position - particles[i])
                )

                particles[i] = particles[i] + velocities[i]
                particles[i] = np.clip(particles[i], self.lower_bounds, self.upper_bounds)

                score = self.fitness(particles[i])

                if score > personal_best_scores[i]:
                    personal_best_scores[i] = score
                    personal_best_positions[i] = particles[i].copy()

                if score > self.global_best_score:
                    self.global_best_score = score
                    self.global_best_position = particles[i].copy()

            self.convergence.append(self.global_best_score)

            print(
                f"Generation {gen + 1}/{self.generations} | "
                f"Best PSO Accuracy: {self.global_best_score:.4f}"
            )

        best_params = self.decode_position(self.global_best_position)
        return best_params, self.global_best_score, self.convergence


pso = PSOOptimizer(
    population_size=20,
    generations=25,
    inertia=0.7,
    cognitive=1.5,
    social=1.5
)

best_params, best_pso_accuracy, pso_convergence = pso.optimize()

print("\nBest PSO Parameters:")
print(best_params)
print("Best PSO Accuracy:", best_pso_accuracy)


# ============================================================
# 10. TRAIN PSO OPTIMIZED CLASSIFIER
# ============================================================

pso_model = RandomForestClassifier(
    n_estimators=best_params["n_estimators"],
    max_depth=best_params["max_depth"],
    min_samples_split=best_params["min_samples_split"],
    min_samples_leaf=best_params["min_samples_leaf"],
    random_state=42
)

pso_model.fit(X_train, y_train)

pso_pred = pso_model.predict(X_test)

pso_accuracy = accuracy_score(y_test, pso_pred)
pso_precision = precision_score(y_test, pso_pred, average="weighted", zero_division=0)
pso_recall = recall_score(y_test, pso_pred, average="weighted", zero_division=0)
pso_f1 = f1_score(y_test, pso_pred, average="weighted", zero_division=0)

results.append({
    "Model": "PSO Optimized Random Forest",
    "Accuracy": pso_accuracy,
    "Precision": pso_precision,
    "Recall": pso_recall,
    "F1_Score": pso_f1
})

predictions["PSO Optimized Random Forest"] = pso_pred

results_df = pd.DataFrame(results)

print("\nFinal Model Results:")
print(results_df)


# ============================================================
# 11. REGRESSION MODEL
# ============================================================

reg_model = RandomForestRegressor(
    n_estimators=best_params["n_estimators"],
    max_depth=best_params["max_depth"],
    min_samples_split=best_params["min_samples_split"],
    min_samples_leaf=best_params["min_samples_leaf"],
    random_state=42
)

reg_model.fit(X_train_reg, y_train_reg)

reg_pred = reg_model.predict(X_test_reg)

regression_metrics = {
    "MAE": float(mean_absolute_error(y_test_reg, reg_pred)),
    "MSE": float(mean_squared_error(y_test_reg, reg_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_test_reg, reg_pred))),
    "R2_Score": float(r2_score(y_test_reg, reg_pred))
}

print("\nRegression Results:")
print(regression_metrics)


# ============================================================
# 12. ANN MODEL FOR H5
# ============================================================

num_classes = len(np.unique(y))

ann_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

ann_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=15,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_loss, ann_accuracy = ann_model.evaluate(X_test, y_test, verbose=0)

results_df = pd.concat([
    results_df,
    pd.DataFrame([{
        "Model": "ANN",
        "Accuracy": ann_accuracy,
        "Precision": np.nan,
        "Recall": np.nan,
        "F1_Score": np.nan
    }])
], ignore_index=True)


# ============================================================
# 13. SAVE MODELS
# ============================================================

h5_path = os.path.join(BASE_DIR, "pso_regional_office_accessibility_model.h5")
pkl_path = os.path.join(BASE_DIR, "pso_regional_office_accessibility_model.pkl")
yaml_path = os.path.join(BASE_DIR, "pso_regional_office_accessibility_config.yaml")
json_path = os.path.join(BASE_DIR, "pso_regional_office_accessibility_metadata.json")

ann_model.save(h5_path)

model_package = {
    "algorithm": "Particle Swarm Optimization",
    "pso_best_parameters": best_params,
    "pso_best_accuracy": best_pso_accuracy,
    "pso_model": pso_model,
    "base_models": base_models,
    "regression_model": reg_model,
    "scaler": scaler,
    "label_encoders": label_encoders,
    "target_encoder": target_encoder,
    "feature_columns": feature_cols
}

with open(pkl_path, "wb") as f:
    pickle.dump(model_package, f)


config = {
    "project_name": "Smart Regional Office Accessibility System",
    "algorithm": "Particle Swarm Optimization",
    "dataset_path": DATASET_PATH,
    "base_directory": BASE_DIR,
    "target_column": target_col,
    "regression_target": reg_target_col,
    "feature_columns": feature_cols,
    "pso_best_parameters": best_params,
    "pso_best_accuracy": float(best_pso_accuracy),
    "h5_model": h5_path,
    "pkl_model": pkl_path
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f)


metadata = {
    "project_name": "Smart Regional Office Accessibility System",
    "algorithm": "Particle Swarm Optimization",
    "total_records": int(df.shape[0]),
    "total_columns": int(df.shape[1]),
    "pso_accuracy": float(pso_accuracy),
    "pso_precision": float(pso_precision),
    "pso_recall": float(pso_recall),
    "pso_f1_score": float(pso_f1),
    "ann_accuracy": float(ann_accuracy),
    "ann_loss": float(ann_loss),
    "regression_metrics": regression_metrics,
    "pso_best_parameters": best_params,
    "saved_files": {
        "h5": h5_path,
        "pkl": pkl_path,
        "yaml": yaml_path,
        "json": json_path
    }
}

with open(json_path, "w") as f:
    json.dump(metadata, f, indent=4)


# ============================================================
# 14. SAVE RESULT CSV
# ============================================================

result_csv_path = os.path.join(BASE_DIR, "pso_result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nResult CSV saved:", result_csv_path)


# ============================================================
# 15. SAVE PSO CONVERGENCE CSV
# ============================================================

convergence_df = pd.DataFrame({
    "Generation": np.arange(1, len(pso_convergence) + 1),
    "Best_Accuracy": pso_convergence
})

convergence_csv_path = os.path.join(BASE_DIR, "pso_convergence_result.csv")
convergence_df.to_csv(convergence_csv_path, index=False)

print("PSO convergence CSV saved:", convergence_csv_path)


# ============================================================
# 16. SAVE PREDICTION CSV
# ============================================================

all_predictions_encoded = pso_model.predict(X_scaled)
all_predictions = target_encoder.inverse_transform(all_predictions_encoded)

all_scores = reg_model.predict(X_scaled)

prediction_df = df.copy()
prediction_df["PSO_Predicted_Demand_Category"] = all_predictions
prediction_df["PSO_Predicted_Service_Demand_Score"] = all_scores.round(2)

prediction_csv_path = os.path.join(BASE_DIR, "pso_prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("Prediction CSV saved:", prediction_csv_path)


# ============================================================
# 17. SAVE PROCESSED DATASET
# ============================================================

processed_csv_path = os.path.join(BASE_DIR, "pso_processed_dataset.csv")
df.to_csv(processed_csv_path, index=False)

print("Processed Dataset saved:", processed_csv_path)


# ============================================================
# 18. ACCURACY GRAPH
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(results_df["Model"], results_df["Accuracy"], marker="o")
plt.title("PSO Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.grid(True)
plt.tight_layout()

accuracy_graph_path = os.path.join(BASE_DIR, "pso_accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

print("Accuracy graph saved:", accuracy_graph_path)


# ============================================================
# 19. CONFUSION MATRIX HEATMAP
# ============================================================

cm = confusion_matrix(y_test, pso_pred)

plt.figure(figsize=(7, 5))
plt.imshow(cm)
plt.title("PSO Confusion Matrix Heatmap")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()

class_names = target_encoder.classes_

plt.xticks(np.arange(len(class_names)), class_names)
plt.yticks(np.arange(len(class_names)), class_names)

for i in range(len(cm)):
    for j in range(len(cm[i])):
        plt.text(j, i, cm[i][j], ha="center", va="center")

plt.tight_layout()

heatmap_path = os.path.join(BASE_DIR, "pso_confusion_matrix_heatmap.png")
plt.savefig(heatmap_path, dpi=300)
plt.show()

print("Heatmap saved:", heatmap_path)


# ============================================================
# 20. COMPARISON GRAPH
# ============================================================

comparison_df = results_df[["Model", "Accuracy", "Precision", "Recall", "F1_Score"]].copy()

x = np.arange(len(comparison_df["Model"]))
width = 0.2

plt.figure(figsize=(11, 6))
plt.bar(x - width * 1.5, comparison_df["Accuracy"], width, label="Accuracy")
plt.bar(x - width / 2, comparison_df["Precision"].fillna(0), width, label="Precision")
plt.bar(x + width / 2, comparison_df["Recall"].fillna(0), width, label="Recall")
plt.bar(x + width * 1.5, comparison_df["F1_Score"].fillna(0), width, label="F1 Score")

plt.xticks(x, comparison_df["Model"], rotation=30)
plt.title("PSO Model Performance Comparison")
plt.xlabel("Model")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()

comparison_graph_path = os.path.join(BASE_DIR, "pso_comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

print("Comparison graph saved:", comparison_graph_path)


# ============================================================
# 21. RESULT GRAPH
# ============================================================

plt.figure(figsize=(8, 5))
plt.bar(results_df["Model"], results_df["Accuracy"])
plt.title("PSO Result Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.tight_layout()

result_graph_path = os.path.join(BASE_DIR, "pso_result_graph.png")
plt.savefig(result_graph_path, dpi=300)
plt.show()

print("Result graph saved:", result_graph_path)


# ============================================================
# 22. PREDICTION GRAPH
# ============================================================

category_counts = prediction_df["PSO_Predicted_Demand_Category"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(category_counts.index, category_counts.values)
plt.title("PSO Prediction Graph - Demand Category Count")
plt.xlabel("Predicted Demand Category")
plt.ylabel("Number of Offices")
plt.tight_layout()

prediction_graph_path = os.path.join(BASE_DIR, "pso_prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

print("Prediction graph saved:", prediction_graph_path)


# ============================================================
# 23. SERVICE DEMAND SCORE GRAPH
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(
    prediction_df.index,
    prediction_df["PSO_Predicted_Service_Demand_Score"],
    marker="o"
)
plt.title("PSO Predicted Service Demand Score Graph")
plt.xlabel("Office Index")
plt.ylabel("Predicted Service Demand Score")
plt.grid(True)
plt.tight_layout()

score_graph_path = os.path.join(BASE_DIR, "pso_service_demand_score_graph.png")
plt.savefig(score_graph_path, dpi=300)
plt.show()

print("Service demand score graph saved:", score_graph_path)


# ============================================================
# 24. PSO CONVERGENCE GRAPH
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(convergence_df["Generation"], convergence_df["Best_Accuracy"], marker="o")
plt.title("PSO Convergence Graph")
plt.xlabel("Generation")
plt.ylabel("Best Accuracy")
plt.grid(True)
plt.tight_layout()

convergence_graph_path = os.path.join(BASE_DIR, "pso_convergence_graph.png")
plt.savefig(convergence_graph_path, dpi=300)
plt.show()

print("PSO convergence graph saved:", convergence_graph_path)


# ============================================================
# 25. ANN TRAINING ACCURACY GRAPH
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("PSO ANN Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()

ann_accuracy_graph_path = os.path.join(BASE_DIR, "pso_ann_training_accuracy_graph.png")
plt.savefig(ann_accuracy_graph_path, dpi=300)
plt.show()

print("ANN accuracy graph saved:", ann_accuracy_graph_path)


# ============================================================
# 26. ANN LOSS GRAPH
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("PSO ANN Training Loss Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()

ann_loss_graph_path = os.path.join(BASE_DIR, "pso_ann_training_loss_graph.png")
plt.savefig(ann_loss_graph_path, dpi=300)
plt.show()

print("ANN loss graph saved:", ann_loss_graph_path)


# ============================================================
# 27. FEATURE IMPORTANCE GRAPH
# ============================================================

importance_df = pd.DataFrame({
    "Feature": feature_cols,
    "Importance": pso_model.feature_importances_
}).sort_values("Importance", ascending=False)

importance_csv_path = os.path.join(BASE_DIR, "pso_feature_importance.csv")
importance_df.to_csv(importance_csv_path, index=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.title("PSO Feature Importance Graph")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.gca().invert_yaxis()
plt.tight_layout()

feature_graph_path = os.path.join(BASE_DIR, "pso_feature_importance_graph.png")
plt.savefig(feature_graph_path, dpi=300)
plt.show()

print("Feature importance CSV saved:", importance_csv_path)
print("Feature importance graph saved:", feature_graph_path)


# ============================================================
# 28. CLASSIFICATION REPORT CSV
# ============================================================

report_dict = classification_report(
    y_test,
    pso_pred,
    target_names=target_encoder.classes_,
    output_dict=True,
    zero_division=0
)

report_df = pd.DataFrame(report_dict).transpose()

report_csv_path = os.path.join(BASE_DIR, "pso_classification_report.csv")
report_df.to_csv(report_csv_path)

print("Classification report saved:", report_csv_path)


# ============================================================
# 29. FINAL SUMMARY
# ============================================================

print("\n================================================")
print("PSO PROJECT RESULTS GENERATED SUCCESSFULLY")
print("================================================")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("YAML Config:", yaml_path)
print("JSON Metadata:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Processed CSV:", processed_csv_path)
print("Convergence CSV:", convergence_csv_path)
print("Classification Report CSV:", report_csv_path)
print("Feature Importance CSV:", importance_csv_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Heatmap:", heatmap_path)
print("Comparison Graph:", comparison_graph_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Service Demand Score Graph:", score_graph_path)
print("Convergence Graph:", convergence_graph_path)
print("ANN Accuracy Graph:", ann_accuracy_graph_path)
print("ANN Loss Graph:", ann_loss_graph_path)
print("Feature Importance Graph:", feature_graph_path)
print("================================================")
```
