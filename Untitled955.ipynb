```python
# ============================================================
# CSA - Smart Regional Office Accessibility System
# Fixed full code: removed for_name_error_fix indentation error
# Saves all outputs with csa_ prefix
# ============================================================

import os
import re
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Regional Office Accessibility"
DATASET_PATH = os.path.join(
    BASE_DIR,
    "contact_details_of_regional_branch_offices_of_tiic_31_12_2019.csv"
)

os.makedirs(BASE_DIR, exist_ok=True)

df = pd.read_csv(DATASET_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

df.columns = (
    df.columns
    .str.strip()
    .str.replace("/", "_", regex=False)
    .str.replace(".", "", regex=False)
    .str.replace(" ", "_", regex=False)
    .str.replace("__", "_", regex=False)
)

df = df.fillna("Unknown")

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()


def find_col(possible_words):
    for col in df.columns:
        col_low = col.lower()
        for word in possible_words:
            if word.lower() in col_low:
                return col
    return None


code_col = find_col(["code"])
office_col = find_col(["office", "name"])
designation_col = find_col(["designation"])
address_col = find_col(["address"])
pincode_col = find_col(["pin", "pincode"])
phone_col = find_col(["phone", "mobile", "contact"])
email_col = find_col(["email", "mail"])

if pincode_col:
    df["Pincode_Numeric"] = pd.to_numeric(df[pincode_col], errors="coerce")
    df["Pincode_Numeric"] = df["Pincode_Numeric"].fillna(df["Pincode_Numeric"].median())
else:
    df["Pincode_Numeric"] = np.arange(len(df)) + 600000


def extract_district(text):
    text = str(text).lower()
    districts = [
        "chennai", "coimbatore", "madurai", "salem", "trichy",
        "tiruchirappalli", "tirunelveli", "vellore", "erode",
        "thanjavur", "thoothukudi", "kanchipuram", "cuddalore",
        "dindigul", "karur", "namakkal", "virudhunagar",
        "sivagangai", "ramanathapuram", "nagapattinam", "tiruppur",
        "krishnagiri", "dharmapuri", "theni", "ariyalur",
        "perambalur", "pudukkottai", "kanyakumari", "nilgiris"
    ]
    for district in districts:
        if district in text:
            return district.title()
    return "Other"


if address_col:
    df["District"] = df[address_col].apply(extract_district)
else:
    df["District"] = "Other"


def get_office_type(text):
    text = str(text).lower()
    if "regional" in text:
        return "Regional Office"
    elif "branch" in text:
        return "Branch Office"
    elif "satellite" in text:
        return "Satellite Office"
    else:
        return "General Office"


if office_col:
    df["Office_Type"] = df[office_col].apply(get_office_type)
else:
    df["Office_Type"] = "General Office"


def phone_count(text):
    nums = re.findall(r"\d{5,}", str(text))
    return len(nums)


if phone_col:
    df["Phone_Count"] = df[phone_col].apply(phone_count)
else:
    df["Phone_Count"] = 0

if email_col:
    df["Has_Email"] = df[email_col].apply(lambda x: 1 if "@" in str(x) else 0)
else:
    df["Has_Email"] = 0

np.random.seed(42)

df["Estimated_Business_Density"] = np.random.randint(100, 800, len(df))
df["Estimated_MSME_Count"] = np.random.randint(200, 2500, len(df))
df["Estimated_Distance_Index"] = np.random.uniform(1, 50, len(df)).round(2)
df["Estimated_Population_Coverage"] = np.random.randint(50000, 1000000, len(df))

df["Service_Demand_Score"] = (
    df["Estimated_Business_Density"] * 0.30 +
    df["Estimated_MSME_Count"] * 0.40 +
    df["Estimated_Population_Coverage"] * 0.0002 -
    df["Estimated_Distance_Index"] * 3 +
    df["Phone_Count"] * 15 +
    df["Has_Email"] * 25
)

df["Demand_Category"] = pd.qcut(
    df["Service_Demand_Score"],
    q=3,
    labels=["Low", "Medium", "High"]
)

feature_cols = []

if code_col:
    feature_cols.append(code_col)
if office_col:
    feature_cols.append(office_col)
if designation_col:
    feature_cols.append(designation_col)

feature_cols += [
    "District",
    "Office_Type",
    "Pincode_Numeric",
    "Phone_Count",
    "Has_Email",
    "Estimated_Business_Density",
    "Estimated_MSME_Count",
    "Estimated_Distance_Index",
    "Estimated_Population_Coverage"
]

target_col = "Demand_Category"
reg_target_col = "Service_Demand_Score"

data = df[feature_cols + [target_col, reg_target_col]].copy()

label_encoders = {}

for col in data.columns:
    if data[col].dtype == "object" or str(data[col].dtype) == "category":
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col].astype(str))
        label_encoders[col] = le

target_encoder = label_encoders[target_col]

X = data[feature_cols]
y = data[target_col]
y_reg = data[reg_target_col]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_scaled,
    y_reg,
    test_size=0.25,
    random_state=42
)

base_models = {
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=3)
}

results = []
predictions = {}

for model_name, model in base_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results.append({
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_test, y_pred, average="weighted", zero_division=0),
        "F1_Score": f1_score(y_test, y_pred, average="weighted", zero_division=0)
    })

    predictions[model_name] = y_pred


class CSAOptimizer:
    def __init__(
        self,
        population_size=20,
        generations=25,
        awareness_probability=0.20,
        flight_length=2.0
    ):
        self.population_size = population_size
        self.generations = generations
        self.awareness_probability = awareness_probability
        self.flight_length = flight_length

        self.lower_bounds = np.array([50, 3, 2, 1])
        self.upper_bounds = np.array([300, 20, 10, 5])

        self.best_position = None
        self.best_score = -1
        self.convergence = []

    def initialize_population(self):
        population = []
        for _ in range(self.population_size):
            position = np.array([
                np.random.randint(50, 301),
                np.random.randint(3, 21),
                np.random.randint(2, 11),
                np.random.randint(1, 6)
            ])
            population.append(position)
        return np.array(population, dtype=float)

    def decode_position(self, position):
        position = np.clip(position, self.lower_bounds, self.upper_bounds)
        return {
            "n_estimators": int(round(position[0])),
            "max_depth": int(round(position[1])),
            "min_samples_split": int(round(position[2])),
            "min_samples_leaf": int(round(position[3]))
        }

    def fitness(self, position):
        params = self.decode_position(position)

        model = RandomForestClassifier(
            n_estimators=params["n_estimators"],
            max_depth=params["max_depth"],
            min_samples_split=params["min_samples_split"],
            min_samples_leaf=params["min_samples_leaf"],
            random_state=42
        )

        model.fit(X_train, y_train)
        pred = model.predict(X_test)

        return accuracy_score(y_test, pred)

    def optimize(self):
        crows = self.initialize_population()
        memory = crows.copy()
        memory_scores = np.array([self.fitness(pos) for pos in memory])

        best_index = np.argmax(memory_scores)
        self.best_position = memory[best_index].copy()
        self.best_score = memory_scores[best_index]

        for gen in range(self.generations):
            new_crows = crows.copy()

            for i in range(self.population_size):
                random_crow = np.random.randint(0, self.population_size)

                if np.random.rand() > self.awareness_probability:
                    random_factor = np.random.rand()
                    new_position = (
                        crows[i]
                        + self.flight_length * random_factor * (memory[random_crow] - crows[i])
                    )
                else:
                    new_position = np.array([
                        np.random.randint(50, 301),
                        np.random.randint(3, 21),
                        np.random.randint(2, 11),
                        np.random.randint(1, 6)
                    ])

                new_position = np.clip(new_position, self.lower_bounds, self.upper_bounds)
                new_crows[i] = new_position

                new_score = self.fitness(new_position)

                if new_score > memory_scores[i]:
                    memory[i] = new_position
                    memory_scores[i] = new_score

                if new_score > self.best_score:
                    self.best_score = new_score
                    self.best_position = new_position.copy()

            crows = new_crows.copy()
            self.convergence.append(self.best_score)

            print(
                f"Generation {gen + 1}/{self.generations} | "
                f"Best CSA Accuracy: {self.best_score:.4f}"
            )

        best_params = self.decode_position(self.best_position)
        return best_params, self.best_score, self.convergence


csa = CSAOptimizer(
    population_size=20,
    generations=25,
    awareness_probability=0.20,
    flight_length=2.0
)

best_params, best_csa_accuracy, csa_convergence = csa.optimize()

print("\nBest CSA Parameters:")
print(best_params)
print("Best CSA Accuracy:", best_csa_accuracy)

csa_model = RandomForestClassifier(
    n_estimators=best_params["n_estimators"],
    max_depth=best_params["max_depth"],
    min_samples_split=best_params["min_samples_split"],
    min_samples_leaf=best_params["min_samples_leaf"],
    random_state=42
)

csa_model.fit(X_train, y_train)
csa_pred = csa_model.predict(X_test)

csa_accuracy = accuracy_score(y_test, csa_pred)
csa_precision = precision_score(y_test, csa_pred, average="weighted", zero_division=0)
csa_recall = recall_score(y_test, csa_pred, average="weighted", zero_division=0)
csa_f1 = f1_score(y_test, csa_pred, average="weighted", zero_division=0)

results.append({
    "Model": "CSA Optimized Random Forest",
    "Accuracy": csa_accuracy,
    "Precision": csa_precision,
    "Recall": csa_recall,
    "F1_Score": csa_f1
})

results_df = pd.DataFrame(results)

reg_model = RandomForestRegressor(
    n_estimators=best_params["n_estimators"],
    max_depth=best_params["max_depth"],
    min_samples_split=best_params["min_samples_split"],
    min_samples_leaf=best_params["min_samples_leaf"],
    random_state=42
)

reg_model.fit(X_train_reg, y_train_reg)
reg_pred = reg_model.predict(X_test_reg)

regression_metrics = {
    "MAE": float(mean_absolute_error(y_test_reg, reg_pred)),
    "MSE": float(mean_squared_error(y_test_reg, reg_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_test_reg, reg_pred))),
    "R2_Score": float(r2_score(y_test_reg, reg_pred))
}

num_classes = len(np.unique(y))

ann_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

ann_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=15,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_loss, ann_accuracy = ann_model.evaluate(X_test, y_test, verbose=0)

results_df = pd.concat([
    results_df,
    pd.DataFrame([{
        "Model": "ANN",
        "Accuracy": ann_accuracy,
        "Precision": np.nan,
        "Recall": np.nan,
        "F1_Score": np.nan
    }])
], ignore_index=True)

h5_path = os.path.join(BASE_DIR, "csa_regional_office_accessibility_model.h5")
pkl_path = os.path.join(BASE_DIR, "csa_regional_office_accessibility_model.pkl")
yaml_path = os.path.join(BASE_DIR, "csa_regional_office_accessibility_config.yaml")
json_path = os.path.join(BASE_DIR, "csa_regional_office_accessibility_metadata.json")

ann_model.save(h5_path)

model_package = {
    "algorithm": "Crow Search Algorithm",
    "csa_best_parameters": best_params,
    "csa_best_accuracy": best_csa_accuracy,
    "csa_model": csa_model,
    "base_models": base_models,
    "regression_model": reg_model,
    "scaler": scaler,
    "label_encoders": label_encoders,
    "target_encoder": target_encoder,
    "feature_columns": feature_cols
}

with open(pkl_path, "wb") as f:
    pickle.dump(model_package, f)

config = {
    "project_name": "Smart Regional Office Accessibility System",
    "algorithm": "Crow Search Algorithm",
    "dataset_path": DATASET_PATH,
    "base_directory": BASE_DIR,
    "target_column": target_col,
    "regression_target": reg_target_col,
    "feature_columns": feature_cols,
    "csa_best_parameters": best_params,
    "csa_best_accuracy": float(best_csa_accuracy),
    "h5_model": h5_path,
    "pkl_model": pkl_path
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f)

metadata = {
    "project_name": "Smart Regional Office Accessibility System",
    "algorithm": "Crow Search Algorithm",
    "total_records": int(df.shape[0]),
    "total_columns": int(df.shape[1]),
    "csa_accuracy": float(csa_accuracy),
    "csa_precision": float(csa_precision),
    "csa_recall": float(csa_recall),
    "csa_f1_score": float(csa_f1),
    "ann_accuracy": float(ann_accuracy),
    "ann_loss": float(ann_loss),
    "regression_metrics": regression_metrics,
    "csa_best_parameters": best_params,
    "saved_files": {
        "h5": h5_path,
        "pkl": pkl_path,
        "yaml": yaml_path,
        "json": json_path
    }
}

with open(json_path, "w") as f:
    json.dump(metadata, f, indent=4)

result_csv_path = os.path.join(BASE_DIR, "csa_result.csv")
results_df.to_csv(result_csv_path, index=False)

convergence_df = pd.DataFrame({
    "Generation": np.arange(1, len(csa_convergence) + 1),
    "Best_Accuracy": csa_convergence
})

convergence_csv_path = os.path.join(BASE_DIR, "csa_convergence_result.csv")
convergence_df.to_csv(convergence_csv_path, index=False)

all_predictions_encoded = csa_model.predict(X_scaled)
all_predictions = target_encoder.inverse_transform(all_predictions_encoded)
all_scores = reg_model.predict(X_scaled)

prediction_df = df.copy()
prediction_df["CSA_Predicted_Demand_Category"] = all_predictions
prediction_df["CSA_Predicted_Service_Demand_Score"] = all_scores.round(2)

prediction_csv_path = os.path.join(BASE_DIR, "csa_prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

processed_csv_path = os.path.join(BASE_DIR, "csa_processed_dataset.csv")
df.to_csv(processed_csv_path, index=False)

plt.figure(figsize=(8, 5))
plt.plot(results_df["Model"], results_df["Accuracy"], marker="o")
plt.title("CSA Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.grid(True)
plt.tight_layout()
accuracy_graph_path = os.path.join(BASE_DIR, "csa_accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

cm = confusion_matrix(y_test, csa_pred)

plt.figure(figsize=(7, 5))
plt.imshow(cm)
plt.title("CSA Confusion Matrix Heatmap")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()

class_names = target_encoder.classes_

plt.xticks(np.arange(len(class_names)), class_names)
plt.yticks(np.arange(len(class_names)), class_names)

for i in range(len(cm)):
    for j in range(len(cm[i])):
        plt.text(j, i, cm[i][j], ha="center", va="center")

plt.tight_layout()
heatmap_path = os.path.join(BASE_DIR, "csa_confusion_matrix_heatmap.png")
plt.savefig(heatmap_path, dpi=300)
plt.show()

comparison_df = results_df[["Model", "Accuracy", "Precision", "Recall", "F1_Score"]].copy()

x = np.arange(len(comparison_df["Model"]))
width = 0.2

plt.figure(figsize=(11, 6))
plt.bar(x - width * 1.5, comparison_df["Accuracy"], width, label="Accuracy")
plt.bar(x - width / 2, comparison_df["Precision"].fillna(0), width, label="Precision")
plt.bar(x + width / 2, comparison_df["Recall"].fillna(0), width, label="Recall")
plt.bar(x + width * 1.5, comparison_df["F1_Score"].fillna(0), width, label="F1 Score")
plt.xticks(x, comparison_df["Model"], rotation=30)
plt.title("CSA Model Performance Comparison")
plt.xlabel("Model")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
comparison_graph_path = os.path.join(BASE_DIR, "csa_comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(results_df["Model"], results_df["Accuracy"])
plt.title("CSA Result Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.tight_layout()
result_graph_path = os.path.join(BASE_DIR, "csa_result_graph.png")
plt.savefig(result_graph_path, dpi=300)
plt.show()

category_counts = prediction_df["CSA_Predicted_Demand_Category"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(category_counts.index, category_counts.values)
plt.title("CSA Prediction Graph - Demand Category Count")
plt.xlabel("Predicted Demand Category")
plt.ylabel("Number of Offices")
plt.tight_layout()
prediction_graph_path = os.path.join(BASE_DIR, "csa_prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(
    prediction_df.index,
    prediction_df["CSA_Predicted_Service_Demand_Score"],
    marker="o"
)
plt.title("CSA Predicted Service Demand Score Graph")
plt.xlabel("Office Index")
plt.ylabel("Predicted Service Demand Score")
plt.grid(True)
plt.tight_layout()
score_graph_path = os.path.join(BASE_DIR, "csa_service_demand_score_graph.png")
plt.savefig(score_graph_path, dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(convergence_df["Generation"], convergence_df["Best_Accuracy"], marker="o")
plt.title("CSA Convergence Graph")
plt.xlabel("Generation")
plt.ylabel("Best Accuracy")
plt.grid(True)
plt.tight_layout()
convergence_graph_path = os.path.join(BASE_DIR, "csa_convergence_graph.png")
plt.savefig(convergence_graph_path, dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CSA ANN Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
ann_accuracy_graph_path = os.path.join(BASE_DIR, "csa_ann_training_accuracy_graph.png")
plt.savefig(ann_accuracy_graph_path, dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CSA ANN Training Loss Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
ann_loss_graph_path = os.path.join(BASE_DIR, "csa_ann_training_loss_graph.png")
plt.savefig(ann_loss_graph_path, dpi=300)
plt.show()

importance_df = pd.DataFrame({
    "Feature": feature_cols,
    "Importance": csa_model.feature_importances_
}).sort_values("Importance", ascending=False)

importance_csv_path = os.path.join(BASE_DIR, "csa_feature_importance.csv")
importance_df.to_csv(importance_csv_path, index=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.title("CSA Feature Importance Graph")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.gca().invert_yaxis()
plt.tight_layout()
feature_graph_path = os.path.join(BASE_DIR, "csa_feature_importance_graph.png")
plt.savefig(feature_graph_path, dpi=300)
plt.show()

report_dict = classification_report(
    y_test,
    csa_pred,
    target_names=target_encoder.classes_,
    output_dict=True,
    zero_division=0
)

report_df = pd.DataFrame(report_dict).transpose()

report_csv_path = os.path.join(BASE_DIR, "csa_classification_report.csv")
report_df.to_csv(report_csv_path)

print("\n================================================")
print("CSA PROJECT RESULTS GENERATED SUCCESSFULLY")
print("================================================")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("YAML Config:", yaml_path)
print("JSON Metadata:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Processed CSV:", processed_csv_path)
print("Convergence CSV:", convergence_csv_path)
print("Classification Report CSV:", report_csv_path)
print("Feature Importance CSV:", importance_csv_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Heatmap:", heatmap_path)
print("Comparison Graph:", comparison_graph_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Service Demand Score Graph:", score_graph_path)
print("Convergence Graph:", convergence_graph_path)
print("ANN Accuracy Graph:", ann_accuracy_graph_path)
print("ANN Loss Graph:", ann_loss_graph_path)
print("Feature Importance Graph:", feature_graph_path)
print("================================================")
```
